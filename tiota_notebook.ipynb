```python
import matplotlib.pyplot as plt

# Prepare the figure and axes
fig, ax = plt.subplots()
x = Z_dataset_train[0,:,:].flatten()
y = Zt_dataset_train[0,:,:].flatten()
indices = np.arange(len(x))

# Plotting flattened matrices
ax.plot(indices, x, marker='o', color='red', label='Matrix Z')
ax.plot(indices, y, marker='o', color='blue', label='Matrix Zt')

# Adding labels and legend
ax.set_title('Flattened Matrix Values Z and Zt')
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.legend()

# Show the plot
plt.show()
```
